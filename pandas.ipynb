## **Column Operations**

```python
gh['Total'] = gh[['NET','Dev','WeD','PWD','JSC','DVS','EVS']].sum(axis=1) 
gh['Total']
```

```python
gh['Average'] = gh['Total'] / 7  
gh['Average']
```

```python
gh['Result'] = gh['Average'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
gh['Result']
```

```python
gh['Grade'] = pd.cut(gh['Average'], bins=[0,40,60,80,100], labels=['F','C','B','A'])
gh['Grade']
```

```python
gh
```

```python
gh.rename(columns={'NET': 'Network'}, inplace=True) 
```

```python
gh
```

```python
gh['Name'] = gh['Name'].str.upper()  
```

```python
gh
```

```python
gh['Name Length'] = gh['Name'].str.len() 
```

```python
gh[['Name Length']]
```

```python
gh['Rounded Dev'] = gh['Dev'].round(0)
```

```python
gh['Rounded Dev']
```

```python
gh['DVS x 2'] = gh['DVS'] * 2
```

```python
gh['DVS x 2']
```
