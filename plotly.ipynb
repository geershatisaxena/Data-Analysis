```python
import plotly.io as pio

# Set default template globally
pio.templates.default = "plotly_dark"
```

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
mg = {
    'Name': ["Peter", "Quagmire", "Luois", "Stewie", "Brian", "Griffins", "Giant Dog", "Meg", "Obama", "Chris", "Leg", "Putin"],
    'NET': [51, 63, 59, 45, 85, 74, 95, 96, 23, 65, 22.2, 14],
    'Dev': [45, 52, 23, 14, 25, 63, 9, 74, 50, 63.2, 66, 12],
    'WeD': [36, 25, 5, 62, 19, 18, 55, 65, 44, 70, 69, 33.6],
    'PWD': [52, 66, 23, 12, 21, 41, 52, 25, 36, 45.6, 77, 63],
    'JSC': [25, 14, 52, 63, 45, 21, 85, 96, 45, 63, 55.3, 23],
    'DVS': [75, 85, 96, 5, 45, 63, 47, 65, 23.6, 52, 60, 50],
    'EVS': [30, 66, 56, 45, 32.3, 44, 51, 25, 36, 47, 58, 56]
}

gh = pd.DataFrame(mg)
gh
```

## **🌞 SUNBURST CHARTS**

```python
# 1. Simple sunburst
px.sunburst(gh, path=['Name'], values='NET', title="NET Scores")
```

```python
# 2. Sunburst with multiple levels
gh['Level1'] = ['A'] * 6 + ['B'] * 6
px.sunburst(gh, path=['Level1', 'Name'], values='NET', title='Grouped by Level1')
```

```python
# 3. Sunburst using Dev
px.sunburst(gh, path=['Level1', 'Name'], values='Dev', color='Dev')
```

```python
# 4. Sunburst with PWD and color
px.sunburst(gh, path=['Level1', 'Name'], values='PWD', color='PWD', color_continuous_scale='Blues')
```

```python
# 5. JSC score as size, NET as color
px.sunburst(gh, path=['Level1', 'Name'], values='JSC', color='NET', color_continuous_scale='RdBu')
```

```python
# 6. With custom hover data
px.sunburst(gh, path=['Level1', 'Name'], values='DVS', hover_data=['EVS'])
```

```python
# 7. EVS values and labels
px.sunburst(gh, path=['Level1', 'Name'], values='EVS', title="EVS by Group")
```

```python
# 8. Custom depth (only group-level view)
px.sunburst(gh, path=['Level1'], values='NET')
```

```python
# 9. Depth limit 2
px.sunburst(gh, path=['Level1', 'Name'], values='NET', maxdepth=2)
```

```python
# 10. Label sorting
px.sunburst(gh.sort_values('NET'), path=['Level1', 'Name'], values='NET')
```
